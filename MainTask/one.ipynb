```python
# Add 2 minute interval column according to entry_minute for ex, all vecihles entered between 0-2 minutes will have interval 0-2
# all vechicles entered between 2-4 minutes will have interval 2-4 and similarly for all other intervals

# in our case intervals should be like 18:43 - 18:45, 18:45 - 18:47, 18:47 - 18:49 and so on
# for this we can use pd.cut() function

df['Interval'] = pd.cut(df['Entry_Minute'], bins=range(0,61,2), right=False)
df.head()

# add another column for the exit_minute interval
df['Exit_Interval'] = pd.cut(df['Exit_Minute'], bins=range(0,61,2), right=False)
df.head()
```

```python
# Plot the interval column
plt.figure(figsize=(18,6))

#plot exit interval and entry interval
sns.countplot(x='Interval', data=df, color='blue', alpha=0.5)
sns.countplot(x='Exit_Interval', data=df, color='red', alpha=0.5)
plt.title('Number of vehicles in each interval')

# decrease the font size of xticks
plt.xticks(fontsize=8)
plt.show()
```

```python
# Do the seasonal plot for number of vehicles in each interval
plt.figure(figsize=(18,6))
df['Interval'].value_counts().sort_index().plot(style='.-')
df['Exit_Interval'].value_counts().sort_index().plot(style='.-')
plt.title('Number of vehicles in each interval')
plt.show()
```

```python
# Decompose the above plot into trend, seasonality and residual
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df['Interval'].value_counts().sort_index(), model='additive', period=2)
```

```python
# Plot trend seasonality and residual in one plot
plt.figure(figsize=(18,6))
result.trend.plot()
result.seasonal.plot()
result.resid.plot()

# Label the plot
plt.title('Trend, Seasonality and Residual')
plt.legend(['Trend', 'Seasonality', 'Residual'])

plt.show()
```

# Exponential Smoothing (ETS)

```python
from darts.models import ExponentialSmoothing
from darts.utils.utils import ModelMode, SeasonalityMode

# Fit the Simple Exponential Smoothing model
ses_model = ExponentialSmoothing(trend=ModelMode.NONE, seasonal=SeasonalityMode.NONE)
ses_model.fit(series)
ses_forecast = ses_model.predict(15)

# Convert the forecast to Pandas Series and plot
ses_forecast_series = ses_forecast.pd_series()
combined_ses_series = pd.concat([interval_counts, ses_forecast_series])

plt.figure(figsize=(10, 6))
combined_ses_series.plot(label='SES Forecasted', color='orange')
interval_counts.plot(label='Original', color='blue')
plt.legend()
plt.title('Simple Exponential Smoothing Forecast')
plt.xlabel('Interval')
plt.ylabel('Count')
plt.show()
```

```python
# Holt's Exponential Smoothing Model
holtModel = ExponentialSmoothing(trend=ModelMode.ADDITIVE, seasonal=SeasonalityMode.ADDITIVE)
holtModel.fit(series)

holt_forecast = holtModel.predict(15)

holt_forecast_series = holt_forecast.pd_series()
combined_holt_series = pd.concat([interval_counts, holt_forecast_series])

plt.figure(figsize=(10, 6))
combined_holt_series.plot(label='Holt Forecasted', color='orange')
interval_counts.plot(label='Original', color='blue')
plt.legend()
plt.title('Holt Forecast')
plt.xlabel('Interval')
plt.ylabel('Count')
plt.show()
```
